```python
import numpy as np
import matplotlib.pyplot as plt

# Define the Haar wavelet function
def haar_wavelet(x):
    if x >= 0 and x < 0.5:
        return 1
    elif x >= 0.5 and x < 1:
        return -1
    else:
        return 0

# Define the Haar scaling function
def haar_scaling(x):
    if x >= 0 and x < 1:
        return 1
    else:
        return 0

# Vectorize the functions so they can accept numpy arrays
vec_haar_wavelet = np.vectorize(haar_wavelet)
vec_haar_scaling = np.vectorize(haar_scaling)

# Generate a range of x values
x_values = np.linspace(0, 1.2, 100)

# Apply the Haar functions to the range of x values
y_wavelet = vec_haar_wavelet(x_values)
y_scaling = vec_haar_scaling(x_values)

# Plot the Haar wavelet function
plt.subplot(1, 2, 1)
plt.plot(x_values, y_wavelet, drawstyle='steps-post')
plt.title('Haar wavelet function')
plt.ylim(-1.5, 1.5)

# Plot the Haar scaling function
plt.subplot(1, 2, 2)
plt.plot(x_values, y_scaling, drawstyle='steps-post')
plt.title('Approximation of the Haar scaling function')
plt.ylim(-0.2, 1.2)

# Show the plots
plt.show()
```
